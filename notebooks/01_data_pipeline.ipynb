# PRISM Notebook 01: Data Pipeline

Load NFL/NBA play-by-play and market data from DuckDB.

**Prerequisite:** `python scripts/ingest_sports.py` and `python scripts/ingest_markets.py`

```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt

con = duckdb.connect('data/prism.duckdb')
summary = {
    'nfl': con.execute("SELECT COUNT(*) FROM game_states WHERE sport='NFL'").fetchone()[0],
    'nba': con.execute("SELECT COUNT(*) FROM game_states WHERE sport='NBA'").fetchone()[0],
    'matches': con.execute('SELECT COUNT(*) FROM game_contract_map WHERE match_confidence>0.8').fetchone()[0],
}
pd.Series(summary, name='count')
```

```python
game = con.execute("""
  SELECT seconds_remaining, home_score, away_score FROM game_states
  WHERE sport='NFL' ORDER BY game_id, seconds_remaining DESC LIMIT 30
""").df()
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(game.seconds_remaining, game.home_score, label='Home')
ax.plot(game.seconds_remaining, game.away_score, label='Away')
ax.invert_xaxis(); ax.legend(); ax.set_title('NFL Score Timeline'); plt.show()
```
