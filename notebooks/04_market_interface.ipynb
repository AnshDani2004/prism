# PRISM Notebook 04: Market Interface

Plot model probability vs market price over game time.

```python
import matplotlib.pyplot as plt
from src.data.database import PrismDatabase
from src.market.edge import EdgeCalculator
from src.models.inplay_xgb import XGBInPlayModel

db = PrismDatabase()
m = db.query_df('SELECT * FROM game_contract_map WHERE match_confidence>0.8 LIMIT 1').iloc[0]
states = db.query_df('SELECT * FROM game_states')
model = XGBInPlayModel().fit(states, XGBInPlayModel.home_win_outcomes(XGBInPlayModel.final_game_states(states)))
edges = EdgeCalculator(db=db).compute_edge_series(m.game_id, model, m.market_source, m.contract_id)
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(edges.seconds_remaining, edges.model_prob, label='Model')
ax.plot(edges.seconds_remaining, edges.market_price, label='Market')
ax.invert_xaxis(); ax.legend(); plt.show()
```
