# PRISM Notebook 03: Bayesian Online Model

EKF vs particle filter on a single game. Same SMC algorithm family as **bayesian-market-filters**.

```python
from src.data.database import PrismDatabase
from src.models.bayesian_online import BayesianOnlineWinProb

states = PrismDatabase().query_df('SELECT * FROM game_states WHERE sport="NFL" LIMIT 500')
gid = states.game_id.iloc[0]
game = states[states.game_id == gid]
outcomes = BayesianOnlineWinProb.home_win_outcomes(BayesianOnlineWinProb.final_game_states(states))
ekf = BayesianOnlineWinProb(inference='ekf').fit(states, outcomes)
pf = BayesianOnlineWinProb(inference='particle', n_particles=500)
pf.process_noise, pf.observation_noise, pf.base_rate = ekf.process_noise, ekf.observation_noise, ekf.base_rate
print('EKF:', ekf.replay_game(game)[-1:], 'PF:', pf.replay_game(game)[-1:])
```
