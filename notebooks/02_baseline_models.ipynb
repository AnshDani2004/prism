# PRISM Notebook 02: Baseline Models & Calibration

Fit Bradley-Terry and margin models. **ECE** = Expected Calibration Error.

```python
from src.data.database import PrismDatabase
from src.models.bradley_terry import BradleyTerryModel
from src.models.margin_model import MarginRatingModel
from src.models.calibration import CalibrationAnalyzer

states = PrismDatabase().query_df('SELECT * FROM game_states')
final = BradleyTerryModel.final_game_states(states)
outcomes = BradleyTerryModel.home_win_outcomes(final)
bt = BradleyTerryModel().fit(states, outcomes)
mm = MarginRatingModel().fit(states, outcomes)
```

```python
test = final[final.season == 2023]
y = BradleyTerryModel.home_win_outcomes(test).to_numpy()
analyzer = CalibrationAnalyzer()
print('BT ECE:', analyzer.ece(bt.predict(test), y))
print('MM ECE:', analyzer.ece(mm.predict(test), y))
analyzer.plot_reliability_diagram(mm.predict(test), y, 'margin_model')
```
